```python
import csv
names = ['easy', 'medium', 'hard']
with open('puzzles.csv', 'w', newline='') as csvfile:
    for filename in names:
        fieldnames = ['Puzzle', 'Difficulty']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        lines = open(filename+'.txt').read().splitlines()
        for line in lines:
            line = list(line)
            line = ['.' if x == '0' else x for x in line]
            line = ''.join(str(e) for e in line)
            print(len(line))
            writer.writerow({'Puzzle': str(line), 'Difficulty': filename})
```
